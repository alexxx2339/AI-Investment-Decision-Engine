```python
# ============================================
# Section 18 - Trend Detection
# ============================================

latest_ma20 = nvda["MA20"].iloc[-1]

latest_ma50 = nvda["MA50"].iloc[-1]

print("Latest 20-Day MA:", round(latest_ma20, 2))
print("Latest 50-Day MA:", round(latest_ma50, 2))
```

```python
if latest_ma20 > latest_ma50:
    print("Trend Signal: Bullish")
else:
    print("Trend Signal: Bearish")
```

```python
# Interpretation:
# Since the 20-day moving average is below the 50-day moving average,
# the short-term trend is weaker than the medium-term trend.
# This suggests a bearish signal.
```

```python
import numpy as np

nvda["Signal"] = np.where(
    nvda["MA20"] > nvda["MA50"],
    "Bullish",
    "Bearish"
)

print(nvda[["Close", "MA20", "MA50", "Signal"]].tail(10))
```

```python
# ============================================
# Section 20 - Golden Cross / Death Cross
# ============================================

nvda["Previous Signal"] = nvda["Signal"].shift(1)

print(nvda[["Signal", "Previous Signal"]].tail(10))
```

```python
# ============================================
# Section 20 - Detect Trend Changes
# ============================================

nvda["previous Signal"] = nvda["Signal"].shift(1)

print(nvda[["Signal", "Previous Signal"]].tail(10))
```

```python
# ============================================
# Section 21 - Buy/Sell Signals
# ============================================

buy_signals = nvda[
(nvda["Signal"] == "Bullish") & 
(nvda["Previous Signal"] == "Bearish")
]

sell_signals = nvda[
    (nvda["Signal"] == "Bearish") &
    (nvda["Previous Signal"] == "Bullish")
]


print("Buy Signals:", len(buy_signals))
print("Sell Signals:", len(sell_signals))
```

```python
# ============================================
# Section 22 - Buy and Sell Signals Chart
# ============================================

plt.figure(figsize=(14,7))

plt.plot(nvda["Close"], label="Closing Price", linewidth=2)
plt.plot(nvda["MA20"], label="20-Day Moving Average", linewidth=2)
plt.plot(nvda["MA50"], label="50-Day Moving Average", linewidth=2, linestyle="--")

plt.scatter(buy_signals.index, buy_signals["Close"], label="Buy Signal", marker="^", s=100)

plt.scatter(sell_signals.index, sell_signals["Close"], label="Sell Signal", marker="v", s=100)

plt.title("NVIDIA Buy/Sell Signals Based on Moving Average Crossovers")
plt.xlabel("Date")
plt.ylabel("Price ($)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# ============================================
# Section 23 - Create Machine Learning Target
# ============================================

nvda["Tomorrow Close"] = close.shift(-1)

print(nvda[["Close", "Tomorrow Close"]].head(10))
```

```python
# ============================================
# Section 24 - Create Target Variable
# ============================================

nvda["Tomorrow Up"] = (nvda["Tomorrow Close"] > close).astype(int)

print(nvda[["Close", "Tomorrow Close", "Tomorrow Up"]].head(10))
```

```python
# ============================================
# Section 25 - Select Features
# ============================================

features = [
    "Close",
    "Volume",
    "Daily Return",
    "MA20",
    "MA50"
]

X = nvda[features]

y = nvda["Tomorrow Up"]

print(X.head())

print()

print(y.head())
```

```python
# ============================================
# Section 26 - Clean the Data
# ============================================

ml_data = nvda.dropna()

print("Rows before cleaning:", len(nvda))
print("Rows after cleaning :", len(ml_data))
```

```python
# ============================================
# Section 27 - Train/Test Split
# ============================================

from sklearn.model_selection import train_test_split

features = ["Close", "Volume", "Daily Return", "MA20", "MA50"]

X = ml_data[features]

y = ml_data["Tomorrow Up"]

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("Training rows:", len(X_train))
print("Testing rows :", len(X_test))

```

```python
# ============================================
# Section 28 - Logistic Regression Model
# ============================================

from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)


model.fit(X_train, y_train)

print("Model training complete!")
```

```python
# ============================================
# Section 29 - Make Predictions
# ============================================

predictions = model.predict(X_test)

print(predictions[:20])
```

```python
# ============================================
# Section 30 - Model Accuracy
# ============================================

from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, predictions)

print(f"Model Accuracy: {accuracy:.2%}")
```

```python
# ============================================
# Section 31 - Random Forest Model
# ============================================

from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

rf_model.fit(X_train, y_train)

rf_predictions = rf_model.predict(X_test)

from sklearn.metrics import accuracy_score

rf_accuracy = accuracy_score(y_test, rf_predictions)

print(f"Random Forest Accuracy: {rf_accuracy:.2%}")
```

```python
# ============================================
# Section 32 - XGBoost Model
# ============================================

from xgboost import XGBClassifier

xgb_model = XGBClassifier(
    n_estimators=100,
    random_state=42,
    eval_metric="logloss"
)

xgb_model.fit(X_train, y_train)

xgb_predictions = xgb_model.predict(X_test)

from sklearn.metrics import accuracy_score

xgb_accuracy = accuracy_score(y_test, xgb_predictions)

print(f"XGBoost Accuracy: {xgb_accuracy:.2%}")
```
